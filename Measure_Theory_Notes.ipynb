# Basics of Measure Theory

### Motivation

To be able to understand Lebesgue integral, we need to define the notion of _measure_. For example, given an interval on the real line, the measure would be the length of the interval. 

=> Measure theory deals with measures of sets. 

## $\sigma$-Algebra

Given a set $X$, we are familiar with the idea of its **power set** $P(X)$ (the set of all possible subsets of $X$). A $\sigma$-algebra $\mathcal{A}$ is a subset of the power set 
$$\mathcal{A} \subseteq P(X)$$
such that its elements are _measurable_ sets, i.e. these are the sets which we can measure.

**Definition $\sigma$-algebra:**

$\mathcal{A} \subseteq P(X)$ is called a $\sigma$-algebra, if:

- $\phi, X \in \mathcal{A}$
- If $A \in \mathcal{A}$, then $A^c \in \mathcal{A}$
- If $A_i \in \mathcal{A}$, then $\bigcup_{i=1}^{\infty} A_i \in \mathcal{A}$

Hence, any $A \in \mathcal{A}$ is called an $\mathcal{A}$-measurable set.

**Example:**

1. $\mathcal{A} = \{\phi, X\}$
2. $\mathcal{A} = P(X)$

In general, $P(X)$ is hard to obtain, hence the $\sigma$-algebras we deal with typically lie between 1 and 2. 

**Side note:**

For $\mathcal{A}_i$ $\sigma$-algebras on $X$, s.t. $i \in I$ (index set), $\bigcap_{i \in I} \mathcal{A}_i$ is also a $\sigma$-algebra on $X$.

## Smallest $\sigma$-algebra

**Definition:**

For $\mathcal{M} \subseteq P(X)$, there is a smallest $\sigma$-algebra that contains $\mathcal{M}$ (called $\sigma$-algebra generated by $\mathcal{M}$):
$$\sigma(\mathcal{M}) = \bigcap_{A \supseteq \mathcal{M}} \mathcal{A}$$
given $\mathcal{A}$ are $\sigma$-algebras.

**Example:**

Let $X = \{a, b, c, d\}, \quad \mathcal{M} = \{ \{a\}, \{b\} \}$, then 
$$\sigma(\mathcal{M}) = \{\phi, X, \{a\}, \{b, c, d\}, \{b\}, \{a, c, d\}, \{a, b\}, \{c, d\}  \}$$

## Borel $\sigma$-algebra

Let $X$ be a topological space, or a metric space or a subset of $\mathbb{R}^n$ or ... (essentially we need an open set), then a borel $\sigma$-algebra $B(X)$ on $X$ is the $\sigma$-algebra generated by the open sets. 

## What is a Measure?

**Definition:**

Let a set $X$ with its $\sigma$-algebra $\mathcal{A}$, s.t. $(X, \mathcal{A})$ is a measurable space.

A map $\mu: \mathcal{A} \xrightarrow{} [0, \infty]$ is called a measure if it satisfies the following:

- $\mu(\phi) = 0$
- $\mu(\bigcup_{i = 1}^n A_i) = \sum_{i=1}^n \mu(A_i)$ with $A_i \cap A_j = \phi$ for all $A_i \in \mathcal{A}$

=> The triplet $(X, \mathcal{A}, \mu)$ is a measure space.

**Example:**

Given set $X$ and $\mathcal{A} = P(X)$,

1. Counting measure: 

$$ \mu(A)=   \left\{
\begin{array}{ll}
      \text{number of elements in }A & A\text{ is finite} \\
      \infty & \text{else}
\end{array} 
\right.  $$

2. Dirac measure for $p \in X$ (for a point $p$ with concentrated measure):

$$ \delta_p(A)=   \left\{
\begin{array}{ll}
      1 & p \in A \\
      0 & \text{else}
\end{array} 
\right.  $$


### Measure Problem

To find a measure $\mu$ on $P(\mathbb{R})$ s.t.

1. $\mu([a, b]) = b - a$, if $b > a$
2. $\mu(x + A) = \mu(A)$ for $A \in P(\mathbb{R})$ and $x$ is a translation vector

(Such a measure on the power set does not exist)

**Claim:** 

Let $\mu$ be a measure on $P(\mathbb{R})$ with $\mu((0, 1]) < \infty$ and be translation invariant $\Leftrightarrow \mu = 0$.

**Proof:**

Let $I = (0, 1]$ with equivalence relation:
$$x \sim y :\Leftrightarrow x-y \in \mathbb{Q}$$
that is, the equivalence class for an element $x \in I$ is 
$$[x] := \{ x + r | r \in \mathbb{Q}, x+r \in I \}$$ 
for a rational number $r$.

(The quotient set $I/\sim$, which is the set of equivalence classes under the relation $\sim$, can be seen as a partition of $I$ into distinct classes, each consisting of elements that differ by a rational number but possibly uncountably many)

Now, let $A \subseteq I$ s.t.
- For each $[x]$, there is an $a \in A$ ith $a \in [x]$
- For all $a, b \in A: a, b \in [x] \implies a = b$

$A_n := r_n + A$ where $r_n$ is an enumeration of $Q \cap (-1, 1]$

_Claim 1:_ If $A_n \cap A_m \ \phi$, then $n \ne m$

_Proof:_

Let $x \in A_n \cap A_m$. This implies 
$$x = r_n + a_n = r_m + a_m$$
$$a_n - a_m = r_m - r_n \in \mathbb{Q}$$
$$\implies a_n \sim a_m$$
$$\implies a_m, a_n \in [a_m]$$
$$\implies a_n = a_m$$
$$\implies r_m = r_n$$
$$\implies n = m$$

_Claim 2:_ $(0, 1] \subseteq \bigcup A_n \subseteq (-1, 2]$

Now, assume: $\mu$ is a measure on $P(\mathbb{R})$ with $\mu((0, 1]) < \infty$ and translation invariance. 

By the property of translation invariance
$$\mu(r_n + A) = \mu(A) \text{ for all } n \in \mathbb{N}$$

We know that $\mu((0, 1]) =: C < \infty$

$$\mu((-1, 2]) = \mu( (-1, 0] \cup (0, 1] \cup (1, 2] ) = 3C$$

By Claim 2, 
$$\mu((0, 1]) \leq \mu(\bigcup A_n) \leq \mu((-1, 2])$$

$$\implies C \leq \sum_{n=1}^{\infty} \mu(A_n) \leq 3C$$

$$\implies C \leq \sum_{n=1}^{\infty} \mu(A) \leq 3C$$

Only possible if $\mu(A) = 0 \implies \mu = 0$.

## Measurable Maps

**Definition:**

Let two measurable spaces $(\Omega_1, \mathcal{A}_1)$ and $(\Omega_2, \mathcal{A}_2)$, and a map
$$f: \Omega_1 \rightarrow{} \Omega_2$$

$f$ is called a measurable map w.r.t $\mathcal{A}_1$ and $\mathcal{A}_2$ if the preimage of anything in $\mathcal{A}_2$ exists in $\mathcal{A}_1$, that is:
$$f^{-1}(A_2) \in \mathcal{A}_1 \text{ for all } A_2 \in \mathcal{A}_2$$

**Example:**

Given $(\Omega, \mathcal{A})$, $(\mathbb{R}, B(\mathbb{R}))$, a characteristic/indicator function $\chi_A: \Omega \rightarrow{} \mathbb{R}$

$$ \chi_A(\omega)=   \left\{
\begin{array}{ll}
      1 & \omega \in A \\
      0 & \text{else}
\end{array} 
\right.  $$

For all measurable $A \in \mathcal{A}$, $\chi_A$ is a measurable map. 
$$\chi_A^{-1}(\phi) = \phi, \quad \chi_A^{-1}(\mathbb{R}) = \Omega$$
$$\chi_A^{-1}(\{1\}) = A, \quad \chi_A^{-1}(\{0\}) = A^c$$

**Sidenote:** If $f, g$ are measurable, then $g \circ f$ is also measurable.

## Lebesgue Integral

Given a measure space $(X, \mathcal{A}, \mu)$ s.t. $\mu: \mathcal{A} \rightarrow{} [0, \infty]$, and a space $(\mathbb{R}, B(\mathbb{R}))$. We define a measurable map 

$$f: X \rightarrow{} \mathbb{R}$$

$$\implies f^{-1}(E) \in \mathcal{A} \text{ for all Borel sets } E \subseteq \mathbb{R}$$

For example: for the simple indicator function $\chi_A$, the integral is defined as its area:

$$I(\chi_A) = \mu(A)$$

### Simple / Elementary Functions

Any function $f$ is simple if there exist measurable sets $A_1, \cdots, A_n \in \mathcal{A}$ and $c_1, \cdots, c_n \in \mathbb{R}$ such that

$$f(x) = \sum_{i=1}^n c_i \chi_{A_i}(x)$$

Such a simple function is measurable.

**Definition (lebesgue integral):**

Let 
$$\mathcal{S}^+ := \{f: X \rightarrow{} \mathbb{R} | f \text{ is simple, } f \geq 0\}$$

For $f \in \mathcal{S}^+$ with representation $f(x) = \sum_{i=1}^n c_i \chi_{A_i}(x), c_i \geq 0$, the (lebesque) integral of $f$ w.r.t $\mu$ is 

$$I(f) = \int_x f(x) d\mu = \sum_{i=1}^n c_i \mu(A_i) \in [0, \infty]$$

**Properties:**

- $I(\alpha f + \beta g) = \alpha I(f) + \beta I(g)$ for $\alpha, \beta \geq 0$
- If $f \leq g$, then $I(f) \leq I(g)$ (monotonic)
- If $f = g, \mu$-almost everywhere (a.e.), then $\int f d\mu = \int g d\mu$

($\mu$-a.e. means $\mu(\{x \in X | f(x) \ne g(x)\}) = 0$, that is, any changes between $f$ and $g$ happen on 0-measure sets)

- For $f=0$, $I(f) = 0$

**General:**

For a general function $f$, the lebesgue integral of $f$ wrt $\mu$ is

$$I(f) = \int_x f d\mu := \text{sup}\{ I(h) | h \in \mathcal{S}^+, h \leq f \}$$

where $h = \sum c_i \chi_{A_i}$ is the step function.

If $I(f) < \infty$, $f$ is called $\mu$-integrable.

## Monotone Convergence Theorem

Given a measure space $(X, \mathcal{A}, \mu)$, measurable maps $f_n: X \rightarrow{} [0, \infty)$ and $f: X \rightarrow{} [0, \infty)$, s.t. 

$$f_1 \leq f_2 \leq \cdots \quad \mu\text{-a.e}$$
and
$$\text{lim}_{n \rightarrow{} \infty} f_n (x) = f(x) \quad \mu\text{-a.e}$$

$$\implies \text{lim}_{n \rightarrow{} \infty} \int_X f_n (x) d\mu = \int_X f(x) d\mu$$

## Fatou's Lemma

Given a measure space $(X, \mathcal{A}, \mu)$ and measurable maps $f_n: X \rightarrow{} [0, \infty)$,

$$\int_X \text{lim}_{n \rightarrow{} \infty} \text{inf} f_n d\mu \leq \text{lim}_{n \rightarrow{} \infty} \text{inf} \int_X f_n d\mu$$

($\text{inf} f_n$ is also a measurable map. Applying limit and using monotone convergence theorem gives Fatou's lemma)

## Lebesgue's Dominated Convergence Theorem

Given a measure space $(X, \mathcal{A}, \mu)$ and a set of lebesgue integrable functions

$$\mathcal{L}^1(\mu) := \{f: X \rightarrow{} \mathbb{R} \text{ measurable } | \int_X |f|^1 d\mu < \infty\} \quad (\text{L-1 space})$$

For $f \in \mathcal{L}^1(\mu)$, write $f = f^+ - f^-$, then:

$$\int_X f d\mu = \int_X f^+ d\mu - \int_X f^- d\mu$$

**Theorem:** 

Given a sequence of 

$f_n: X \rightarrow{} \mathbb{R}$, measurable for all $n \in \mathbb{N}$ 

and the pointwise limit 

$f: X \rightarrow{} \mathbb{R}$ 

with $f_n \rightarrow{} f$ as $ n \rightarrow{} \infty$ for $x \in X$ ($\mu$-a.e.), 

and $|f_n| \leq g$ with $g \in \mathcal{L}^1(\mu)$ for all $n \in \mathbb{N}$.

($g$ is an integrable majorant i.e. it is greater than all the functions in the sequence)

Then: $f_1, f_2, \cdots \in \mathcal{L}(\mu)$ and $f \in \mathcal{L}(\mu)$, 

$$ \text{lim}_{n \rightarrow{} \infty} \int_X f_n d\mu = \int_X f d\mu $$

## Caratheodory's Extension Theorem

Given a set $X$, a semiring of sets (not necessarily a $\sigma$-algebra) $\mathcal{A} \subseteq P(X)$ and a pre-measure $\mu: \mathcal{A} \rightarrow{} [0, \infty]$

- Then $\mu$ has an extension
$$\hat{\mu}: \sigma(\mathcal{A}) \rightarrow{} [0, \infty], \text{ i.e. } \mu(A) = \hat{\mu}(A) \text{ for all } A \in \mathcal{A}$$

- $\sigma$-finite: If there is a sequence $\{ S_j \}$ with $S_j \in \mathcal{A}, \bigcup_{j=1}^{\infty} S_j = X, \mu(S_j) < \infty$, then the extension $\hat{\mu}$ is unique. 

**Semiring:**

Semiring of sets $\mathcal{A} \subseteq P(X)$

- $\phi \in \mathcal{A}$ (as for $\sigma$-algebras)
- $A, B \in \mathcal{A} \implies A \cap B \in \mathcal{A}$
- For $A, B \in \mathcal{A}$, there are pairwise disjoint sets $S_1, \cdots, S_n \in \mathcal{A}$: $\bigcup_{j=1}^n S_j = A - B$

**Example:**

$\mathcal{A} := \{ [a, b) | a, b \in \mathbb{R}, a \leq b \}$

not a $\sigma$-algebra because $\mathbb{R} \notin \mathcal{A}$ but $\sigma(\mathcal{A}) = B(\mathbb{R})$

- $\phi \in \mathcal{A}$
- 
$$ [a, b) \cap [c, d) =   \left\{
\begin{array}{ll}
      \phi & b \leq c, d \leq a \\
      [c, b) & c \in [a, b), d \notin [a, b) \\
      ...
\end{array} 
\right.  $$
-
$$ [a, b) - [c, d) =   \left\{
\begin{array}{ll}
      [a, b) & b \leq c, d \leq a \\
      [a, c) & c \in [a, b), d \notin [a, b) \\
      [a, c) \cup [d, b)
\end{array} 
\right.  $$

Hence, $\mathcal{A}$ is a semiring.

**Premeasure:**

$\mu: \mathcal{A} \rightarrow{} [0, \infty]$ with $\mathcal{A}$ semiring of sets: 
- $\mu(\phi) = 0$
- $\mu(\bigcup_{j=1}^\infty A_j) = \sum_{j=1}^\infty \mu(A_j)$, for $A \in \mathcal{A}$ and disjoint $A_i$, and $\bigcup A_j \in \mathcal{A}$

**Application:**

$\mathcal{A} := \{ [a, b) | a, b \in \mathbb{R}, a \leq b \}$ is a semiring of sets with $\mu: \mathcal{A} \rightarrow{} [0, \infty]$ and $\mu([a, b)) = b - a$. Then Caratheodory implies there is a unique extension to $B(\mathbb{R})$ which is the lebesgue measure. 

## Lebesgue-Stieltjes measures

(A way to construct a measure)

Given $F: \mathbb{R} \rightarrow{} \mathbb{R}$, a monotonically increasing function. The length of an interval $[a, b)$ is given by $\mu_F = F(b^+) - F(a^+)$

For $\mathcal{A} := \{ [a, b) | a, b \in \mathbb{R}, a \leq b \}$ a semiring of sets, Carathrodory implies there exists exactly one measure $\mu_F: B(\mathbb{R}) \rightarrow{} [0, \infty]$. This is called the _Lebesgue-Stieltjes_ measure for $F$.

**Examples:**

- $F(x) = x, \mu_F = b-a$ (lebesgue measure)
- $F(x) = 1, \mu_F = 0$ (zero measure)
- 
$$ F(x) =   \left\{
\begin{array}{ll}
      0 & x < 0 \\
      1 & x \geq 0
\end{array} 
\right.  $$
then $\mu_F = 1$ (dirac measure)
- $F: \mathbb{R} \rightarrow{} \mathbb{R}$ which is monotonically increasing and continuosuly differentiable, and $F': \mathbb{R} \rightarrow{} [0, \infty]$, $\mu_F = F(b) - F(a) = \int_a^b F'(x)dx$ gives
($\mu_F: A \rightarrow{} \int_A F'(x)dx$)

(Lebesque measure is the unit measure which sends the interval to its normal length)

## Radon-Nikodym Theorem and Lebesgue Decomposition Theorem

Given a measure space $(X, \mathcal{A}, \Delta)$. For this, let $X = \mathbb{R}, \mathcal{A} = B(\mathbb{R}), \Delta = \lambda$ for $\lambda$ being the lebesgue measure (or any other reference measure). 

Take another measure $\mu: B(\mathbb{R}) \rightarrow{} [0, \infty]$. 

- $\mu$ is called absolutely continuous wrt $\lambda$ if $\lambda(A) = 0 \implies \mu(A) = 0$ for all $A \in B(\mathbb{R})$ (this is written as $\mu << \lambda$)

- $\mu$ is called singular if there is $N \in B(\mathbb{R})$ with $\lambda(N) = 0$ and $\mu(N^c) = 0$ (this is written as $\mu \perp \lambda$)

### Decomposition Thm

For $\mu: B(\mathbb{R}) \rightarrow{} [0, \infty]$

- there are two uniquely determined measures $\mu_{ac}, \mu_s: B(\mathbb{R}) \rightarrow{} [0, \infty]$ with $\mu = \mu_{ac} + \mu_s, \mu_{ac} << \lambda, \mu_s \perp \lambda$

### Radon-Nikodym Thm

There is a measurable map $h: \mathbb{R} \rightarrow{} [0, \infty)$ with $\mu_{ac}(A) = \int_A h d\lambda$ for all $A \in B(\mathbb{R})$. Here $h$ is called the density.

## Image Measure

Let two measureable spaces $(X, \mathcal{A})$ and $(Y, \mathcal{C})$, and a measurable map $h: X \rightarrow{} Y$. 
Let the measure of former be $\mu$. Using this can we measure subsets on the latter space using some measure $\hat{\mu}$?

Yes! The measure for the latter space, called an image measure, is
$$\hat{\mu}(c) = \mu(h^{-1} (c))$$

This $\hat{\mu}$ can also be written as $h_* \mu$ or $\mu \circ h^{-1}$.

## Substitution Rule

In the above scenario, let $g: Y \rightarrow{} \mathbb{R}$, then

$$\int_Y g(y) d(h_* \mu)(y) = \int_X g(h(x)) d\mu(x) $$

## Product Measure and Cavalieri's Principle

Given two measure spaces $(X_1, \mathcal{A}_1, \mu_1)$ and $(X_2, \mathcal{A}_2, \mu_2)$, a product measure $\mu$ for the product space $(X_1 \times X_2, \mathcal{A}, \mu)$ can be defined. 

### Product Measure

For subsets $A_1 \in \mathcal{A}_1$ and $A_2 \in \mathcal{A}_2$, the product measure $\mu(A_1 \times A_2)$ should be equal to the product of their individual measures. 
$$\mu(A_1 \times A_2) = \mu_1(A_1)\mu_2(A_2)$$

This measure is general is not unique.

### Product $\sigma$-algebra

Set of rectangles ($\mathcal{A}_1 \times \mathcal{A}_2$) are not necessarily a $\sigma$-algebra, but a semiring of sets. So for the product $\sigma$-algebra, we need the $\sigma$-algebra corresponding to the product. 

$$\mathcal{A} = \sigma(\mathcal{A}_1 \times \mathcal{A}_2)$$

### Cavalieri's Principle

**Proposition:** If $\mu_1$ and $\mu_2$ are $\sigma$-finite, then there is exactly one measure $\mu$ which satisfies above. 

Let set $M \subseteq \mathcal{A}_1 \times \mathcal{A}_2$. Then the measure satisfies:
$$\mu(M) = \int_{X_2} \mu_1(M_y) d\mu_2(y) = \int_{X_1} \mu_2(M_x) d\mu_1(x)\$$

**Example:** Calculate the volume (lebesgue measure in $\mathbb{R}^3$ = product measure construction with lebesque measures on $\mathbb{R}$) of the pyramid with corners $(-1, -1, 0), (-1, 1, 0), (1, -1, 0), (1, 1, 0), (0, 0, 1)$.

The pyramid is a subset $K \subset \mathbb{R}^3$ such that 
$$K = \{(x, y, z) \in \mathbb{R}^3 | 0 \leq z \leq 1, |x| \leq 1-z, |y| \leq 1-z\}$$

If the volume of the pyramid is $\mu(K)$, it is the product measure of $\mu_1$ and $\mu_2$ with $\mu_1$ being a lebesque measure in $\mathbb{R}$ on the z-coordinate, and $\mu_2$ being a lebesque measure on $\mathbb{R}^2$ on the x-y coordinate, s.t.

$$\mu(K) = \int_\mathbb{R} \mu_2(M_z) d\mu_1(z)$$
where $M_z = \{(x, y) \in \mathbb{R}^2 | |x|\leq 1-z, |y|\leq 1-z\}$

(imagine you're looking at level sets--squares--of pyramid from the top)

$$\implies \mu(K) = \int_{[0, 1]} 4(1-z)^2 d\mu_1(z)$$
$$= \int_0^1 4(1-z)^2 dz = 4/3$$

## Fubini's Theorem

Given $\\sigma$-finite $\mu_1$ and $\mu_2$, with their unique product measure $\mu$ and measurable map $f: X_1 \times X_2 \rightarrow{} [0, \infty]$, then
$$\int_{X_1 \times X_2} f d\mu = \int_{X_2} (\int_{X_1} f(x, y) d\mu_1 (x)) d\mu_2 (y) = \int_{X_1} (\int_{X_2} f(x, y) d\mu_2 (y)) d\mu_1 (x)$$

**Example:** Let $\mu$ be a lebesgue measure for $\mathbb{R}^2$, $A = \{(x, y) \in [0, 1] \times [0, 1] | x\geq y \geq x^2\}$ and $f(x, y) = 2xy$. Then the area:

$$\int_A f d\mu = \int_{\mathbb{R}^2} f \chi_A d\mu = \int_{\mathbb{R}} (\int_\mathbb{R} f(x, y) \chi_A(x, y) dy) dx$$

$$\int_A f d\mu = \int_0^1 (\int_{x^2}^x 2xy dy) dx = 2\int_0^1 x (\int_{x^2}^x y dy)dx$$

$$\int_A f d\mu = 2\int_0^1 x\frac{1}{2} (x^2 - x^4) dx = \int_0^1 (x^3 - x^4)$$


## Outer Measures

(not quite needed unless you want to prove Caratheodory's extension theorem)
(also not an attribute for a measure and not necessarily a measure)

**Definition:** A map $\psi: P(X) \rightarrow{} [0, \infty]$ is called an outer measure if

- $\psi(\phi) = 0$
- $A \subseteq B \implies \psi(A) \leq \psi(B)$ (monotonic)
- $A_1, A_2, \cdots \in P(X) \implies \psi(\bigcup_{n=1}^\infty A_n) \leq \sum_{n=1}^\infty \psi(A_n)$

**Question:** Given a $\psi$, how to get a $\mu$?

**Definition ($\psi$-measurable):** Let $\psi$ be an outer measure. $A \in P(X)$ is called $psi$-measurable if for all $Q \in P(X)$ we have:
$$\psi(Q) = \psi(Q \cap A) + \psi(Q \cap A^c)$$

**Proposition:** If $\psi: P(X) \rightarrow{} [0, \infty]$ is an outer measure, then:
- $\mathcal{A}_\psi = \{A \subseteq X | A \text{ is }\psi\text{-measurable}\}$ is a $\sigma$-algebra
- $\mu: \mathcal{A}_\psi \rightarrow{} [0, \infty], \mu(A) = \psi(A)$ is a measure

## Riemann vs Lebesgue Integral

### Riemann

$f: \mathbb{R} \rightarrow{} \mathbb{R}$

- Difficult to expand to higher dimensions
- Depends on continuity of the function
- Limit processes (riemann integral is not flexible to easily take the limit inside the integral $\text{lim}_{n \rightarrow{} \infty} \int_a^b f_n (x) dx = \int_a^b \text{lim}_{n \rightarrow{} \infty} f_n (x) dx$)

Decompose the domain and take the sum of the rectangles

### Lebesgue

$f: \Omega \rightarrow{} \mathbb{R}$

Decomposes the range, instead of the domain, to get sets in the domain. And then get a weighted sum of the sets to get the integral.
